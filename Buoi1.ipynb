```
#Bai 10
x = tf.random.uniform(shape=[5], minval=0, maxval=10, dtype=tf.int32)
y = tf.random.uniform(shape=[5], minval=0, maxval=10, dtype=tf.int32)
result = tf.where(x < y, x + y, x - y)
print("x:", x.numpy())
print("y:", y.numpy())
print("result:", result.numpy())
```

```
#Bai 11
x = tf.random.uniform(shape=[3,2], minval=5, maxval=10, dtype=tf.int32)
y = tf.random.uniform(shape=[2,3], minval=15, maxval=20, dtype=tf.int32)
diag_mask = tf.constant([
    [1,0],
    [0,1],
    [0,0]
], dtype=tf.int32)
x_diag = x * diag_mask
y_resized = tf.transpose(y)
result = x_diag + y_resized
print("x:", x.numpy())
print("y:", y.numpy())
print("x sau khi giu đuong cheo:\n", x_diag.numpy())
print("result:\n", result.numpy())
```
